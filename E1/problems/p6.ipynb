### Train OLS model to get residuals

```python
def add_1s(x):
    Xs = [x, np.ones_like(x)]
    x = np.vstack(Xs).T

    return x
```

```python
B = 100000
N = len(df)

list_slope, list_intercept = [], []


for _ in range(B):
    df_sample = df.sample(N, replace=True)

    X = add_1s(df_sample['x'].to_numpy())

    model = sm.OLS(df_sample['y'], X)
    results = model.fit()

    list_slope.append(results.params['x1'])
    list_intercept.append(results.params['const'])
```

```python
print(f"Boostrap: slope    : {np.std(list_slope):.5f}")
print(f"Boostrap: intercept: {np.std(list_intercept):.5f}")
```
