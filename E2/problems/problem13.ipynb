```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
```

```python
counts = [
    [[15, 8]],
    [[8, 0], [7, 8]],
    [[8, 0], [0, 7], [7, 1]],
    [[8, 0], [0, 7], [0, 1], [7, 0]],
]
```

```python
results = []

for n, c in enumerate(counts):
    impurity = []
    for i, reg in enumerate(c):
        ratio1, ratio2 = reg[0] / sum(reg), reg[1] / sum(reg)

        gini = ratio1 * (1 - ratio1) + ratio2 * (1 - ratio2)

        impurity.append(gini * sum(reg) / 23)

    print(f"Split: {n + 1} ==> impurity: {sum(impurity) ** 2:.4f}")
```

```python
pd.DataFrame.from_records(results)
```
