## Examine processed data

```python
path = "data/data_merged_int.xlsx"

df_proc = pd.read_excel(path)
```

```python
df_proc.head()
```

```python
df_af = df_proc[df_proc['Entity'] == 'Afghanistan']
```

```python
s1, s2 = 'gdp_growth', 'gdp_per_capita'
t1, t2, t3 = 'Electricity from renewables (TWh)', 'Capacity(MW)', 'Investment(USDmn)'

def pearson(df):
    return {
        'corr_gdpg_erenewable': np.corrcoef(df[s1], df[t1])[0, 1],
        'corr_gdpg_capacity': np.corrcoef(df[s1], df[t2])[0, 1],
        'corr_gdpg_investment': np.corrcoef(df[s1], df[t3])[0, 1],
        'corr_gdp_erenewable': np.corrcoef(df[s2], df[t1])[0, 1],
        'corr_gdp_capacity': np.corrcoef(df[s2], df[t2])[0, 1],
        'corr_gdp_investment': np.corrcoef(df[s2], df[t3])[0, 1]
    }

s_pearson = df_proc.groupby('Entity').apply(pearson)
```

```python
df_pearson = pd.DataFrame(s_pearson.values.tolist(), index=s_pearson.index)
```
