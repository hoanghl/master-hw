```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
```

```python
plt.style.use('seaborn-v0_8')
plt.rcParams.update({'font.size': 8})
```

```python
path = "ex4_p3_times.csv"

df = pd.read_csv(path)
df.head()
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)

sns.lineplot(df, x='n', y='time', hue='flag', ax=ax)
ax.set_ylabel("Time (second)")
ax.set_xlabel("Num. unrolled commands")
```
